```python
from brian2 import *
%matplotlib inline
```

## FitzHugh-Nagumo

```python
start_scope()

tau = 10*ms

eqs = Equations('dv/dt = (((v*(0.1*v)*(v-1))-w)+0.5)/(10*ms):1 ')
eqs += Equations('dw/dt = ((0.8*v)-(0.5*w))/(10*ms):1 ')

G = NeuronGroup(1, eqs, threshold='v>0.3', reset='v = 0', refractory=5*ms, method='euler')

statemon = StateMonitor(G, 'v',record=0)

run(100*ms)

plot(statemon.t/ms, statemon.v[0])
xlabel('Time (ms)')
ylabel('v');

```

## 8.2.2 Intrinsically Bursting (IB) Neurons

```python
start_scope()

tau = 10*ms

eqs = Equations('dv/dt = (((v+75)*(v+45))-u+0.1)/(150*tau):1 ')
eqs += Equations('du/dt = (0.01*(5*(v+75)-u))/tau:1 ')

G = NeuronGroup(1, eqs, threshold='v>-45', reset='v = -75', method='euler')
G.v= -75
statemon = StateMonitor(G, 'v',record=0)
spikemon = SpikeMonitor(G)

run(100*ms)

plot(statemon.t/ms, statemon.v[0])
for t in spikemon.t:
    axvline(t/ms, ls='--', c='C1', lw=3)
xlabel('Time (ms)')
ylabel('mv');
```

## 8.2.3 Multi-Compartment Dendritic Tree

```python
start_scope()

tau = 10*ms

eqs = Equations('dv/dt = ((3*(v+0.06)*(v+0.05))-u+0.5)/(100*tau):1 ')
eqs += Equations('du/dt = (0.01*(5*(v+0.06)-u))/tau:1 ')

G = NeuronGroup(1, eqs, threshold='v>-0.05', reset='v = -0.06', refractory=5*ms, method='rk4')
G.v=-0.06
statemon = StateMonitor(G, 'v',record=0)
spikemon = SpikeMonitor(G)

run(200*ms)

plot(statemon.t/ms, statemon.v[0])
for t in spikemon.t:
    axvline(t/ms, ls='--', c='C1', lw=3)
xlabel('Time (ms)')
ylabel('v');
```

## 8.2.4 Chattering (CH) Neurons

```python
start_scope()

tau = 10*ms

eqs = Equations('dv/dt = ((1.5*(v+0.06)*(v+0.04))-u+0.5)/(50*tau):1 ')
eqs += Equations('du/dt = (0.03*((v+0.06)-u))/tau:1 ')

G = NeuronGroup(1, eqs, threshold='v>-0.04', reset='v = -0.06', refractory=5*ms, method='rk4')
G.v=-0.06
statemon = StateMonitor(G, 'v',record=0)
spikemon = SpikeMonitor(G)

run(100*ms)

plot(statemon.t/ms, statemon.v[0])
for t in spikemon.t:
    axvline(t/ms, ls='--', c='C1', lw=3)
xlabel('Time (ms)')
ylabel('v');
```

## 8.2.6 Fast Spiking (FS) Interneurons

```python
start_scope()

tau = 10*ms

eqs = Equations('dv/dt = ((v+55)*(v+40)-u+0.5)/(20*tau):1 ')
eqs += Equations('du/dt = (0.02*(-u))/tau:1 ')

G = NeuronGroup(1, eqs, threshold='v>-40', reset='v = -55', refractory=10*ms, method='euler')
G.v=-55
statemon = StateMonitor(G, 'v',record=0)
spikemon = SpikeMonitor(G)

run(200*ms)

plot(statemon.t/ms, statemon.v[0])
for t in spikemon.t:
    axvline(t/ms, ls='--', c='C1', lw=3)
xlabel('Time (ms)')
ylabel('v');
```
